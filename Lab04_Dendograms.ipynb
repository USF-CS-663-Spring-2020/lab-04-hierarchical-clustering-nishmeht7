```
import pandas as pd

df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```

```
# Dendogram for NIPS dataset
df = pd.read_csv("NIPS_1987-2015.csv")
df
```

```
isinstance(df, pd.DataFrame)
```

```
# inorder to run linkage we can't use non numeric values, so dropped unnamed column
df = df.drop(["Unnamed: 0"], axis = 1)
df
```

```
Z = linkage(df, 'ward')
```

```
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```
